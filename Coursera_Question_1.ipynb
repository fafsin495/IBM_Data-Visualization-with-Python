```python
import pandas as pd 
```

```python
crime_df = pd.read_csv("https://cocl.us/datascience_survey_data", index_col = 0)
```

```python
crime_df.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
crime_df_percent = ((crime_df / 2233) * 100).round(2)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
bar_plot = crime_df_percent.plot(kind='bar',
                       figsize = (20, 8),
                       width = 0.8,
                       color = ['#5cb85c', '#5bc0de', '#d9534f'],
                       fontsize = 14)

plt.title('Percentage of Respondents Interests\''' in Data Science Areas', fontsize=16)
bar_plot.set_facecolor((1.0,1.0,1.0))

y_pivot = bar_plot.axes.get_yaxis()
y_pivot.set_visible(False)

[bar_plot.annotate(str(i.get_height()) + '%', (i.get_x() * 1.005, i.get_height() * 1.03))   for i in bar_plot.patches]
```
